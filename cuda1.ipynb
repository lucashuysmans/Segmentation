```python
import torch
import ClassFiles.networks as net
import ClassFiles.networks2 as net2
import ClassFiles.GeneratedDataset as dat
import ClassFiles.Training as train
import ClassFiles.DeepSegmentation as rec
from torch.utils.data import DataLoader
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
train_clean_image_dataset = dat.ImageDataset("/nfs/st01/hpc-cmih-cbs31/lh568/data/train", "clean")
train_cv_image_dataset = dat.ImageDataset("/nfs/st01/hpc-cmih-cbs31/lh568/data/train", "chan-vese")

train_clean_image_dataloader = DataLoader(train_clean_image_dataset, batch_size=100, pin_memory = True, shuffle = True)
train_cv_image_dataloader = DataLoader(train_cv_image_dataset, batch_size=100, pin_memory = True, shuffle = True)
```

```python
NN = net2.ConvNet8(1, 128, 128)
NN.load_state_dict(torch.load('ConvNet8_trained_evenmore'))

train.train_regulariser(NN, train_clean_image_dataloader, train_cv_image_dataloader,
                       epochs = 100, lr = 0.00007, mu = 15, binary = True, flip = False, device = "cuda:1",
                       save_path = "cuda1_tempnet")
```

```python
train.train_regulariser(NN, train_clean_image_dataloader, train_cv_image_dataloader,
                       epochs = 100, lr = 0.00001, mu = 15, binary = True, flip = False, device = "cuda:1",
                       save_path = "cuda1_tempnet_2")
```

```python
train.train_regulariser(NN, train_clean_image_dataloader, train_cv_image_dataloader,
                       epochs = 100, lr = 0.000001, mu = 15, binary = True, flip = False, device = "cuda:1",
                       save_path = "cuda1_tempnet_3")
```

```python
clean_image_dataset = dat.ImageDataset("./Neural_Networks_lunglike/eval", "clean")
dirty_image_dataset = dat.ImageDataset("./Neural_Networks_lunglike/eval", "dirty")
cv_seg_dataset = dat.SegmentationDataset("./Neural_Networks_lunglike/eval", "chan-vese")



image_number = 1

NN = net2.ConvNet6(1, 128, 128)
NN.load_state_dict(torch.load("./cuda1_tempnet"))

dirty_im = Image.fromarray(
    255 * dirty_image_dataset.__getitem__(image_number)[0].numpy()
).convert("L")
cv_seg = cv_seg_dataset.__getitem__(image_number)

reconstruction = rec.DeepSegmentation(dirty_im, NN, cv_seg)
reconstruction.show_segmentation()
reconstruction.run(1000, lmb_reg=20, epsilon=0.01, show_iterations=True)
reconstruction.show_segmentation()
```
