```python
import ClassFiles.GeneratedDataset as dat
```

```python
clean_dataset = dat.ImageDataset('./Neural_Networks_lunglike/eval', 'clean')
dirty_dataset = dat.ImageDataset('./Neural_Networks_lunglike/eval', 'dirty')
cv_dataset = dat.ImageDataset('./Neural_Networks_lunglike/eval', 'chan-vese')
```

```python
from torch.utils.data import DataLoader
clean_dataloader = DataLoader(clean_dataset, batch_size = 2)
dirty_dataloader = DataLoader(dirty_dataset, batch_size = 2)
cv_dataloader = DataLoader(cv_dataset, batch_size = 2)
```

```python
import torch
import ClassFiles.networks as net
NN = net.ConvNet1(1, 128, 128)
NN.load_state_dict(torch.load('./Neural_Networks_lunglike/ConvNet1_trained'))
```

```python
from Neural_Networks.Code.algorithm2_old import reconstruct, shitty_reconstruct

clean_batch = iter(clean_dataloader).next()
cv_batch = iter(cv_dataloader).next()
"""
make initial image binary
"""
cv_batch = (cv_batch > 0.5).float()
dirty_batch = iter(dirty_dataloader).next()

reconstructed_batch, _ = reconstruct(dirty_batch, dirty_batch, NN, epsilon = 0.5, reconstruction_steps = 1000, lambda_reg = 0.3, lambda_data = 1000)

reconstructed_batch_numpy = reconstructed_batch.to("cpu").numpy()
reconstructed_batch_numpy_squeeze = reconstructed_batch_numpy.squeeze()

from PIL import Image
for i in range(2):
    im = Image.fromarray(255*reconstructed_batch_numpy_squeeze[i]).convert("L")
    im.save(fp = "./Neural_Networks_lunglike/reconstructed_ConvNet1/reconstructed_"+str(i).zfill(2)+".png", format = 'PNG')
```
