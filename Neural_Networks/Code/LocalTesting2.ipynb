```python
import torch
"""
initialise the desired neural network from networks.py
will have to include the PATH to the networks.py file
"""
from 'PATH' import ConvNet1
NN = ConvNet1(1, 128, 128)
NN.to("cuda:0") #change to gpu (get rid of this if no gpu, or change to cuda:1 if want to use a second gpu)

"""
load an existing NN if required
(note you MUST initialise an NN of the SAME KIND before loading the parameters for it)
"""
load = False
if load == True:
    NN.load_state_dict(torch.load('PATH TO NN FILE LOCATION'))
```

```python
"""
training using the file algorithm1_testing:

saves to PATH,
epochs = 20,
batchsize = 100,
shuffle = True,
(not shuffling means it is trained on pairs of corresponding chanvese, groundtruth images)
binary = True,
(changes images to 0 for white, 1 for black, threshold is automatically set at 0.5 inside algorithm1_testing)

within algorithm1_testing it will automatically subtract -0.5 to make black 0.5, white -0.5

flip = False,
(fliping would randomly change values to minus themselves, so white (-0.5) becomes 0.5, and black (0.5) becomes -0.5)
(flipping is not done to pairs of chanvese, groundtruth images independently, see algorithm1_testing for details)

device = "cuda:0",
(change to "cpu" if no gpu, or to "cuda:1" if want to use a second gpu)
"""
from 'PATH TO algorithm1_testing' import train
NN = train(NN, 'PATH TO save location for NN', epochs = 20, batch_size = 100 , mu = 5, shuffle = True, binary = True, flip = False, l2_gradients = True, device = "cuda:0")

"""
output values:

at end of each epoch, and before any training outputs 4 values: (mean is taken over the 100 evaluation images):
'mean wasserstein distance (value on groundtruth - value on chanvese),
max value of NN on groundtruth segmentations,
min value of NN on chanvese segmentations,
maximum of (wasserstein distance on coordinated pairs of groundtruth, chanvese images in evaluation set) (WANT TO BE SMALL)
proportion classified INCORRECTLY (value on groundtruth is larger than value of chanvese)'

during each batch, outputs 2 values: (mean is taken over the batch_size images in the training batch):
mean wasserstein distance (value on groundtruth - value on chanvese),
mean gradient loss (how large the term enforcing the lipschitz condition is: RULE OF THUMB, ADJUST mu UNTIL VALUE FLOATS AROUND 1.0)
"""
```

```python
"""
reconstruct some images

currently using algorithm2_old, this should all be redone with Peter's chanvese code
"""
import torch
from 'PATH TO DATALOADER' import get_generated_dataloader
from 'PATH TO algorithm2_old' import reconstruct, shitty_reconstruct

"""
device to do the computations on, change accordingly
"""
device = "cuda:0"

"""
load the images to reconstruct into dataloader
WILL HAVE TO CHANGE SINCE DATALOADER NO LONGER WORKS THE SAME WAY, AND SINCE IMAGES ARE NO LONGER STORED IN THE SAME LOCATION

trick: set batch_size = 100 to get all evaluation images in one go
"""
eval_groundtruth_loader = get_generated_dataloader('eval', 'clean', batch_size = 100, shuffle = False)
eval_chanvese_loader = get_generated_dataloader('eval', 'chan-vese', batch_size = 100, shuffle = False)
eval_noisy_loader = get_generated_dataloader('eval', 'dirty', batch_size = 100, shuffle = False)

eval_groundtruth_batch = iter(eval_groundtruth_loader).next()[0].to(device)
eval_chanvese_batch = iter(eval_chanvese_loader).next()[0].to(device)
"""
make initial image binary
"""
eval_chanvese_batch = (eval_chanvese_batch > 0.5).float()
eval_noisy_batch = iter(eval_noisy_loader).next()[0].to(device)
"""
the - 0.5 (before feeding to the NN) is done inside the reconstruction algorithm
THIS WILL HAVE TO BE ADDED IF NOT USING algorithm2_old
"""

NN.to(device)
"""
chose shitty_reconstruct (gradient descent along NN with no datafitting term),
or reconstruct (gradient descent along sum of datafitting term and NN term)
"""
shitty_reconstruct = True
if shitty_reconstruct == True:
    
    reconstructed_batch, _ = shitty_reconstruct(eval_chanvese_batch, eval_noisy_batch, NN, epsilon = 0.1, reconstruction_steps = 150)
    
    """
    at end, print:
    average value of NN on reconstructed batch,
    average value of NN on original batch,
    average value of NN on chanvese batch

    this is to give some idea of how long the run the reconstruction for
    """
    print(NN(reconstructed_batch - 0.5).mean().item(), NN(eval_groundtruth_batch - 0.5).mean().item(), NN(eval_chanvese_batch - 0.5).mean().item())

else:
    reconstructed_batch, _ = reconstruct(eval_chanvese_batch, eval_noisy_batch, NN, epsilon = 0.1, reconstruction_steps = 80, lambda_data = 100, lambda_reg = 0.05)
    """
    while reconstructing, each reconstruction step will print: (mean is taken over batch of images we are reconstructing)
    mean of datafitting term (before multiplying by lambda_data),
    standard deviation of datafitting term (before multiplying by lambda_data),
    mean of regularising term (before multiplying by lambda_reg),
    standard deviation of regularising term (before multiplying by lambda_reg)
    """

"""
save reconstructed images (saves as reconstructed_00.png, _01.png, _02.png, etc.)
"""
reconstructed_batch_numpy = reconstructed_batch.to("cpu").numpy()
reconstructed_batch_numpy_squeeze = reconstructed_batch_numpy.squeeze()

from PIL import Image
for i in range(100):
    im = Image.fromarray(255*reconstructed_batch_numpy_squeeze[i]).convert("L")
    im.save(fp = "PATH TO FOLDER TO SAVE IN/reconstructed_"+str(i).zfill(2)+".png", format = 'PNG')
```
